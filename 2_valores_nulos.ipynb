```python
# Cargamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# cargar archivo en el data frame data
# se despliegan las cinco primeras registros de la BD
data=pd.read_csv('inmuebles2.csv')
data.head(5)
```

```python
# Al observarla se observa que tiene una fila, la primera que se debe borrar
# y tambien dos columnas innecesarias, la 1 y la 2.
# Para eliminar la primera fila que tiene nombres repetidos de los campos
data1=pd.read_csv('inmuebles2.csv',skiprows=1)
data1.head(5)
```

```python
#eliminar las dos primeras columnas
# axis = 1 para columnas
# axis = 0 para filas
data2=data1.drop(['0','Referencia'], axis=1)
data2.head(5)
```

```python
# identificar datos nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
# identificar el total de los valores nulos
nulos_total=data2.isnull().sum().sum()
nulos_total
```

Metodos de sustituccion de valores nulos

```python
# metodo 1, media o promedio
# lo aplicaremos a la variable numerica superficie
# pero antes se hara una copia del dataframe 2 a uno 3
data3=data2.copy()
```

```python
# Metodo 1: media o promedio
# variable numerica superficie
data3['Superficie'] = data3['Superficie'].fillna(data3['Superficie'].mean())
```

```python
# identificar datos nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
# Metodo 2: mediana
# lo aplicaremos a la variable numerica Precio Venta
data3['Precio Venta'] = data3['Precio Venta'].fillna(data3['Precio Venta'].median())
```

```python
# identificar datos nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
# Método 3: sustituir por un valor especifico "10"
# Este código encuentra los blancos o valores faltantes
# de la variable Fecha Alta y los rellena con el valor "10"
data3["Fecha Alta"]=data3["Fecha Alta"].fillna(10)
```

```python
# identificar datos nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
# Cuarto metodo de sustitucion de valores
# sustituir valores nulos por un string concreto
data3["Fecha Venta"]=data3["Fecha Venta"].fillna("2030-06-19 00:00:00")
data3
```

```python
# identificar datos nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
# 5 metodo
# sustituir VN por VN hacia adelante "forward fill" ("ffill")
#filtro por columnas
data3["Tipo"]=data3["Tipo"].fillna(method="ffill")
```

```python
# identificar datos nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#sexto metodo
# Este metodo mira por el valor no faltante mas cercano despues del valor faltante
# y usa ese valor para llenar el vacio.Este proceso se repite para todos los
#valores faltantes en la misma columna Operacion
data3["Operación"]=data3["Operación"].fillna(method="bfill")
```

```python
# identificar datos nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
# Metodo 7:
# Imagine the "Provincia" column has a missing value. This line of code looks back up
# the column to find the closest previous value that's not missing and uses that value
# to replace the missing one. It essentially "carries forward" the last known value to
# fill the gap.
data3["Provincia"]=data3["Provincia"].fillna(method="ffill")
```

```python
# identificar datos nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data3
```

```python
# Metodo 8:
# cambiar una variable string de valor nulo por un valor especifico
# por ejemplo "generico"
data3["Vendedor"]=data3["Vendedor"].fillna("generico")
```

```python
#listar los valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
# Ahora realizamos una copia del dataframe data2 a un dataframe data4
data4=data2.copy()
data4
```

Metodo de eliminacion de los nuelos por columnas

```python
# Este metodo elimina todas las filas donde haya nulos
# Esto significa que la BD se reduce en tamaño, se pierden datos
data4=data2.dropna()
data4
```

```python
# identificar datos nulos
valores_nulos=data4.isnull().sum()
valores_nulos
```
