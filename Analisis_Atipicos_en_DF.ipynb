```python
#Primer paso:
# Se importa siempre las librerías necesarias para realizar el procesamiento estadístico
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ejemplo de creación de un DataFrame para ilustrar
data = {
    'Columna1': [10, 12, 12, 13, 12, 14, 15, 16, 100],  # Incluyendo un outlier
    'Columna2': [20, 21, 22, 21, 23, 24, 25, 30, 200]   # Incluyendo un outlier
}
df2 = pd.DataFrame(data)
```

```python
df2
```

```python
# Calcular Q1 (25%) y Q3 (75%)
Q1 = df2['Columna1'].quantile(0.25)
Q3 = df2['Columna1'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Definir límites para identificar outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
```

```python
# Identificar outliers
outliers = df2[(df2['Columna1'] < limite_inferior) | (df2['Columna1'] > limite_superior)]

print("Outliers encontrados:")
print(outliers)
```

```python
# Retirar los outliers del DataFrame
df2_sin_outliers = df2[~((df2['Columna1'] < limite_inferior) | (df2['Columna1'] > limite_superior))]

print("\nDataFrame sin outliers:")
print(df2_sin_outliers)
```

```python
# Visualización antes y después de eliminar outliers
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x=df2['Columna1'])
plt.title('Con Outliers')

plt.subplot(1, 2, 2)
sns.boxplot(x=df2_sin_outliers['Columna1'])
plt.title('Sin Outliers')

plt.show()
```

```python
#Quinto paso:
#Declaramos las variables dependientes e independientes para la regresión lineal
Vars_Indep= df2_sin_outliers[['D','C','U']]
#Vars_Indep= df2[['D']]
Var_Dep= df2_sin_outliers['S']
```

```python
#Sexto paso:
#Se define model como la función de regresión lineal
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
#Ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep, y=Var_Dep)
```

```python
#Octavo paso:
#Verificamos los coeficientes obtenidos para el modelo ajustado, pidiendo el diccionario del #modelo
model.__dict__
```

```python
#Doceavo paso:
#Corroboramos cual es el coeficiente de Determinación de nuestro modelo
coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)
coef_Deter
```

```python
#Treceavo paso:
#Corroboramos cual es el coeficiente de Correlación de nuestro modelo
coef_Correl=np.sqrt(coef_Deter)
coef_Correl
```
