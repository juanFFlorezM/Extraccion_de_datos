```python
# Cargamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# cargar archivo listings en el data frame data
# se despliegan las cinco primeras registros de la BD
data=pd.read_csv('listings.csv')
data.head(5)
```

```python
# La base de datos esta sin problema de duplicidad de indices y de nombres de campos
# se decide listar la cantidad de nulos en cada variable
data.isnull().sum()
```

```python
# se decide calcular la cantidad de valores nulos en toda la base de datos
data.isnull().sum().sum()
```

```python
# se decide imprimir solo las variables que contienen valores nulos
data.isnull().any()
```

```python
# ya que la BD tiene 75 variables solo imprimire las variables
# que tienen un numero de nulos mayor de cero
null_counts=data.isnull().sum()
print(null_counts[null_counts>0])
```

```python
#Se averigua el tipo de las variables que tienen valores nulos
#mayores que cero en el dataframe data
null_cols=null_counts[null_counts>0].index
data[null_cols].dtypes
```

```python
# se hace una copia del dataframe data a un dataframe data2:
data2=data.copy()
```

```python
# Este codigo permite sustituir los valores nulos en las variables tipo float64 del dataframe data2,
# que tengan valores nulos mayores de cero, por el valor promedio de cada variable:
#primer paso:
# Calcula la cantidad de nulos por columna
null_counts = data2.isnull().sum()
#segundo paso:
# Filtra para obtener solo las columnas con nulos mayores a cero
null_counts = null_counts[null_counts > 0]

# Itera sobre las columnas del DataFrame
for col in data2.columns:
    # Verifica si la columna tiene tipo 'float64' y tiene valores nulos
    if data2[col].dtype == 'float64' and col in null_counts.index:
        # Rellena los valores nulos con la media de la columna
        data2[col].fillna(data2[col].mean(), inplace=True)
#
```

```python
#Se averigua el tipo de las variables que tienen valores nulos
#mayores que cero en el dataframe data2:
null_counts=data2.isnull().sum()
print(null_counts[null_counts>0])
```
